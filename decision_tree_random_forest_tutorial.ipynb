# Decision Trees and Random Forests Tutorial
Welcome! In this notebook, we'll explore the logic, implementation, and evaluation of Decision Trees and Random Forests using the Breast Cancer dataset.

## Step 1: Import Required Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix
```

## Step 2: Load and Explore the Dataset

```python
data = load_breast_cancer()
X = data.data
y = data.target

df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y
df.head()
```

## Step 3: Dataset Summary
We'll inspect some basic statistics about the dataset.

```python
df.describe()
```

## ‚öñÔ∏è Step 4: Data Scaling and Splitting
We'll scale the features and split the dataset into training and testing sets.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

## üå≥ Step 5: Train a Decision Tree Classifier

```python
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
y_pred_tree = tree_clf.predict(X_test)
```

## Step 6: Train a Random Forest Classifier

```python
forest_clf = RandomForestClassifier(n_estimators=100, random_state=42)
forest_clf.fit(X_train, y_train)
y_pred_forest = forest_clf.predict(X_test)
```

## Step 7: Evaluate Both Models

```python
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred_tree))

print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_forest))
```

## Step 8: Confusion Matrix Visualization

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(confusion_matrix(y_test, y_pred_tree), annot=True, fmt='d', cmap='Oranges', ax=ax[0])
ax[0].set_title('Decision Tree Confusion Matrix')
ax[0].set_xlabel('Predicted')
ax[0].set_ylabel('Actual')

sns.heatmap(confusion_matrix(y_test, y_pred_forest), annot=True, fmt='d', cmap='Greens', ax=ax[1])
ax[1].set_title('Random Forest Confusion Matrix')
ax[1].set_xlabel('Predicted')
ax[1].set_ylabel('Actual')

plt.tight_layout()
plt.show()
```

## Step 9: Visualize a Simplified Decision Tree

```python
plt.figure(figsize=(20, 10))
plot_tree(tree_clf, filled=True, feature_names=data.feature_names, class_names=data.target_names, max_depth=3)
plt.title("Decision Tree (max_depth=3)")
plt.show()
```

## ‚úÖ Summary
- Decision Trees are easy to interpret but may overfit.
- Random Forests reduce variance by aggregating predictions from multiple trees.
- Random Forests generally outperform single Decision Trees on unseen data.
